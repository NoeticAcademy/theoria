# Databases - Relational Databases

---

An RDBMS (Relational Database Management System) is software that stores and manages data as collections of tables with rows and columns.


## Characteristics

* Tabular (Rows and Columns)
* SQL (Structured Query Language) for data manipulation.
* ACID transactions

## Keys

### Primary Key

* Uniquely identifies each row in a table.
* Must contain unique values and cannot be null.

### Foreign Key
* A column (or group of columns) in one table that references the primary key in another table.
* Enforces referential integrity (the foreign key value must exist in the referenced table, i.e. there can't be orphan values).
* Data Validation: They restrict the data that can be inserted or updated in a referencing table, ensuring consistency.


## Schema

A schema is the formal definition of the structure of a database.

* Tables: Names, column names, and data types (e.g., integer, text, date).
* Constraints: Rules such as primary keys, foreign keys, unique constraints, and NOT NULL.
* Relationships: How tables relate through keys and constraints.

## Index

Lookup table acting on a schema to speed up SQL queries.

## View

A view is a virtual table based on the result of a query.

Use cases:
* Simplify complex queries: Encapsulate complicated joins, filters, or calculations into a single object.
* Security: Restrict access by exposing only specific columns or rows without granting access to the base tables.
* Immutability: views are *typically* read-only.
