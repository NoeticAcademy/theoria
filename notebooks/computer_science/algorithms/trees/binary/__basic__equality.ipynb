# Implementation

```python
from typing import Optional

from importnb import Notebook

from theoria.validor import TestCase, Validor

with Notebook():
    from notebooks.computer_science.algorithms.trees.binary.__basic__binary_search_tree import (
        BST,
        Node,
    )
    from notebooks.computer_science.algorithms.trees.binary.__basic__insertion import (
        insert_multiple,
    )
```

```python
def bst_equal(t1: BST, t2: BST) -> bool:
    def _bst_equal(n1: Optional[Node], n2: Optional[Node]) -> bool:
        if n1 is None and n2 is None:
            return True
        if n1 is None or n2 is None:
            return False
        if n1.value != n2.value:
            return False
        return _bst_equal(n1.left, n2.left) and _bst_equal(n1.right, n2.right)

    return _bst_equal(t1.root, t2.root)
```

# Tests

```python
test_cases = [
    TestCase(
        input_data={"t1": BST(), "t2": BST()},
        expected_output=True,
        description="Both BSTs are empty",
    ),
    TestCase(
        input_data={
            "t1": insert_multiple(BST(), [4, 2, 7, 1, 3]),
            "t2": insert_multiple(BST(), [4, 2, 7, 1, 3]),
        },
        expected_output=True,
        description="Equal BSTs with same structure and values",
    ),
    TestCase(
        input_data={
            "t1": insert_multiple(BST(), [4, 2, 7, 1, 3]),
            "t2": insert_multiple(BST(), [4, 2, 7, 1, 5]),
        },
        expected_output=False,
        description="BSTs with same structure but different values",
    ),
    TestCase(
        input_data={
            "t1": insert_multiple(BST(), [4, 2, 7, 1]),
            "t2": insert_multiple(BST(), [4, 2, 7, 1, 3]),
        },
        expected_output=False,
        description="BSTs with different structures",
    ),
]
```

```python
if __name__ == "__main__":
    Validor(bst_equal).add_cases(test_cases).run()
```
