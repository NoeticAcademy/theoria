# Binary Search Trees

---

Tree where every node satisfies: `left < root < right` (recursively)


```bash
    8 
   / \
  3   10
 / \    \
1   6    14
   / \
  4   7
```

# Implementation

````python
from dataclasses import dataclass
from typing import Any, Optional
````

````python
# Deliberately implemented here instead of .py - easier to reference all notebooks w.r.t non-utility code.
@dataclass
class Node:
    value: Any
    left: Optional["Node"] = None
    right: Optional["Node"] = None


class BST:
    def __init__(self):
        self.root: Optional[Node] = None
        self.size: int = 0
````

# Complexity

Time (insertion, search, deletion, ...):

* Average $O (\log n)$, 
* Worst $O(n)$ - unbalanced trees - imagine left heavy tree $[1, 2, 3, ..., n]$.
