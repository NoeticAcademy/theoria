## Binary Search

---

Find the index of the desired `target` from a sorted `array` of length $n$.

## Example 1:
```python
Inputs:
array = [1, 3, 5]
target = 3

Output:
1
```

## Example 2:
```python
Inputs:
array = [1, 3, 5]
target = 4

Output:
-1 # (not found)
```

# Implementation

````python
from typing import List

from theoria.validor import TestCase, Validor
````

````python
class BinarySearch:
    def __call__(self, array: List[int], target: int) -> int:
        left, right = 0, len(array) - 1
        while left <= right:
            mid = left + (right - left) // 2
            if array[mid] == target:
                return mid
            elif array[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return -1
````

# Tests

````python
test_cases = [
    TestCase(
        input_data={"array": [], "target": 1},
        expected_output=-1,
        description="Empty array",
    ),
    TestCase(
        input_data={"array": [1, 2, 3, 4, 5], "target": 3},
        expected_output=2,
        description="Target present in the middle of the array",
    ),
    TestCase(
        input_data={"array": [1, 2, 3, 4, 5], "target": 1},
        expected_output=0,
        description="Target is the first element",
    ),
    TestCase(
        input_data={"array": [1, 2, 3, 4, 5], "target": 5},
        expected_output=4,
        description="Target is the last element",
    ),
    TestCase(
        input_data={"array": [1, 2, 3, 4, 5], "target": 6},
        expected_output=-1,
        description="Target not present in the array",
    ),
    TestCase(
        input_data={"array": [1, 3, 5, 7, 9, 11, 13, 15], "target": 11},
        expected_output=5,
        description="Target present in a larger array",
    ),
]
````

````python
Validor(BinarySearch()).add_cases(test_cases).run()
````
