## Big O

Upper bound for function growth (w.r.t operations performed) in terms of input size $n$.

$$
f(n) = O(g(n))
$$

if there exists $c$ and $n_0$ such that for all $n \geq n_0$, 

$$|f(n)| \leq c |g(n)|$$

----

### Simple example

For a list $L = \left[l_1, l_2, ..., l_{2n}\right]$, how many operations does it take to find the sum of the odd indexed elements:

$$
l_1 + l_3 + ... + l_{2n - 1} 
$$

* In principle require $\frac{n}{2}$ operations
* Recall $\frac{n}{2} \leq n$

So we can pick $c = 1$ and $g(n) := n$ and call this a linear time algorithm of $O(n)$. 

---

### Recurrence relations

Binary search of a sorted listed involves:
* Each time finding the half that the target belongs to
* and checking equivalence

Thus it admits a recurrence relation:

$$
T(n) = T\left(\frac{n}{2}\right) + O(1)
$$

Expanding:
$$
T(n) = T\left(\frac{n}{2^k}\right) + kO(1)
$$

Expanding further to the base case, which will happen when $n = 2^k$, or equivalently, at $k = \log_2 n$:

$$
T(n) = T\left(1\right) + \log_2 n * O(1) = T(1) + O(\log_2 n)
$$

Since $T(1)$ represents the base case where we have one element in the array, it takes $O(1)$ time to check if our target is in the array. Thus:

$$
T(n) = O(1) + O(\log_2 n) = O(\log_2 n)
$$
