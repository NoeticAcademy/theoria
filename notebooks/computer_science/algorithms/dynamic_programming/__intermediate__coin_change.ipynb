# Dynamic Programming - Coin Change 

--- 

The coin change problem asks for the fewest number of coins needed to make an exact amount using given denominations. Return -1 if impossible. 

````python
from typing import Dict, List, Optional

from theoria.validor import TestCase, Validor
````

````python
def coin_change_recursion(coins: List[int], amount: int) -> int:
    if amount == 0:
        return 0
    if amount < 0:
        return -1

    min_coins = float("inf")
    for coin in coins:
        # Essentially DFS on the decision tree
        res = coin_change_recursion(coins, amount - coin)
        if res >= 0 and res < min_coins:
            min_coins = 1 + res
    return -1 if min_coins == float("inf") else min_coins
````

````python
def coin_change_memo(
    coins: List[int], amount: int, memo: Optional[Dict[int, int]] = None
) -> int:
    if memo is None:
        memo = {}

    if amount in memo:
        return memo[amount]

    if amount == 0:
        return 0
    if amount < 0:
        return -1

    min_coins = float("inf")
    for coin in coins:
        res = coin_change_memo(coins, amount - coin, memo)
        if res >= 0 and res < min_coins:
            min_coins = 1 + res

    memo[amount] = -1 if min_coins == float("inf") else min_coins
    return memo[amount]
````

````python
def coin_change_tabulation(coins: List[int], amount: int) -> int:
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0  # Base case

    for a in range(1, amount + 1):
        for coin in coins:
            if a - coin >= 0:
                dp[a] = min(dp[a], dp[a - coin] + 1)

    return -1 if dp[amount] == float("inf") else dp[amount]
````

````python
test_cases = [
    TestCase(
        input_data={"coins": [1, 2, 5], "amount": 11},
        expected_output=3,
        description="Standard case with multiple coin denominations",
    ),
    TestCase(
        input_data={"coins": [2], "amount": 3},
        expected_output=-1,
        description="Impossible case where amount cannot be formed",
    ),
    TestCase(
        input_data={"coins": [1], "amount": 0},
        expected_output=0,
        description="Edge case with zero amount",
    ),
    TestCase(
        input_data={"coins": [1], "amount": 1},
        expected_output=1,
        description="Edge case with amount equal to smallest coin",
    ),
    TestCase(
        input_data={"coins": [1], "amount": 2},
        expected_output=2,
        description="Edge case with amount as multiple of smallest coin",
    ),
]

Validor(coin_change_recursion).add_cases(test_cases).run()
Validor(coin_change_memo).add_cases(test_cases).run()
Validor(coin_change_tabulation).add_cases(test_cases).run()
````
