# Dynamic Programming - Concepts

---

Dynamic Programming is an algorithmic technique to solve problems exhibiting optimal substructure and overlapping subproblems by:

* Breaking the problem into subproblems
* Solving each subproblem once
* Storing results in a table or cache
* Using computed answers for larger problem solutions

## Properties

Two conditions need to be satisifed:

**Optimal Substructure**: The optimal solution to a problem can be constructed by combining optimal solutions to its subproblems. If you can break a problem into subproblems whose optimal solutions piece together perfectly to solve the bigger problem, the problem has optimal substructure.

Example: If we denote the shortest path of $u \to v$ by $P(u, v)$, the shortest path from node $u \to v$ can broken down as the shortest path from $u \to x$ and then $x \to v$, meaning
$$P(u, v) = P(u, x) + P(x, v)$$

**Overlapping Subproblems**: The same subproblems appear multiple times in the recursive breakdown. Instead of recalculating these subproblems repeatedly (like in naive recursion), store their solutions and reuse them, dropping the runtime dramatically.

Example: To calculate $\text{fib}(5)$, you also need to calculate $\text{fib}(3)$ multiple times. Cache it.
