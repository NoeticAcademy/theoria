```python
from typing import List, Callable, Any


class MergeSort:
    def __init__(self, key: Callable[[...], Any] = lambda x: x):
        self.key = key

    def __call__(self, array: List[Any]) -> List[Any]:
        return self._sort(array)

    # Divide and Conquer: sort and merge
    def _sort(self, array: List[Any]) -> List[Any]:
        if len(array) <= 1:
            return array

        mid = len(array) // 2
        left = self._sort(array[:mid])
        right = self._sort(array[mid:])
        return self._merge(left, right)

    def _merge(self, left: List[Any], right: List[Any]) -> List[Any]:
        sorted_array = []

        # Two pointers to track positions in left and right arrays
        i = j = 0

        while i < len(left) and j < len(right):
            if self.key(left[i]) <= self.key(right[j]):
                sorted_array.append(left[i])
                i += 1
            else:
                sorted_array.append(right[j])
                j += 1

        return sorted_array
```

```python
data = [3, 6, 8, 10, 1, 2, 1]
MergeSort()(data)
```
